```python
from sklearn.datasets import fetch_openml
```

```python
mnist = fetch_openml('mnist_784', version=1)
##从 OpenML 数据库中下载 MNIST 数据集
##下载的数据集存储于 $HOME/scikit_learn_data 目录中
```

```python
mnist.keys()
```

```python
X, y = mnist.data, mnist.target
##定义数据集中的数据类与目标结果类
a, b, c, d, e, f, g = mnist.frame, mnist.categories, mnist.feature_names, \
mnist.target_names, mnist.DESCR, mnist.details, mnist.url
##定义数据集中的其他类
```

```python
print("X 的类型是：", type(X))  ##Pandas 的 DataFrame 类型
print("y 的类型是：", type(y))
print("a 的类型是：", type(a))
print("b 的类型是：", type(b))
print("c 的类型是：", type(c))
print("d 的类型是：", type(d))
print("e 的类型是：", type(e))
print("f 的类型是：", type(f))
print("g 的类型是：", type(g))
```

```python
%%capture  ##魔法命令：捕获并隐藏单元格的输出，包括标准输出和标准错误。
import matplotlib as mlp
import matplotlib.pyplot as plt

def save_show_img(num):  ##自定义函数：用于生成并保存数据集中的图片
  some_digist = X.values[num]
  some_digist_image = some_digist.reshape(28, 28)

  plt.imshow(some_digist_image, cmap="binary")
  plt.axis("off")
  plt.savefig(f"images/{num}_plot.png", format='png')  ##花括号做变量名保护
  plt.show()  ##plt.show() 会清空当前的绘图缓冲区，导致保存的图片为空。因此需放在 plt.savefig() 之后运行。

save_show_img(9)  ##生成第9个样本的图片

#for serial in range(10):
#  save_show_img(serial)
##批量生成图片
```
